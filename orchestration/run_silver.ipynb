# Silver Layer Orchestration

## What this does
This notebook orchestrates the full **Silver layer** in :contentReference[oaicite:0]{index=0}.

It sequentially triggers all Silver transformation notebooks using `dbutils.notebook.run`.

Each child notebook receives:
mode = "parallel"

(so they can internally decide whether to run parallel logic or incremental logic).

## Execution order
The notebooks run **one by one** in this exact sequence:

1. silver_crm_cust_info
2. silver_crm_prd_info
3. silver_crm_sales_details
4. silver_erp_cust_az12
5. silver_erp_loc_a101
6. silver_erp_px_cat_g1v2

## Behavior
- Sequential execution (dependency-safe)
- Fail-fast (pipeline stops on first failure)
- No timeout
- Parameters passed to every notebook

## Implementation
```python
dbutils.notebook.run(path, timeout_seconds=0, arguments={"mode": "parallel"})

````python
silver_notebooks = [
    "/notebooks/silver/silver_crm_cust_info",
    "/notebooks/silver/silver_crm_prd_info",
    "/notebooks/silver/silver_crm_sales_details",
    "/notebooks/silver/silver_erp_cust_az12",
    "/notebooks/silver/silver_erp_loc_a101",
    "/notebooks/silver/silver_erp_px_cat_g1v2"
]

for nb in silver_notebooks:
    print(f"Running {nb}")
    dbutils.notebook.run(nb, timeout_seconds=0,arguments={"mode": "parallel"})
````
